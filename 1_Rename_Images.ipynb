```python
import cv2
import os
import mahotas
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from skimage.io import imread, imshow
```

```python
drive.mount('/content/drive')
data_folder = '/content/drive/My Drive/Group/images/'
```

```python
labels = os.listdir(data_folder)
for label in labels:
    label_folder = os.path.join(data_folder, label)

    # Get the list of images in the label folder
    images = os.listdir(label_folder)

    # Loop through the images and rename them
    for i, image_name in enumerate(images):
        image_path = os.path.join(label_folder, image_name)

        # Define the new image name in the format "{label}_{i}.jpg"
        new_image_name = f"{label}_{i + 1}.jpg"
        new_image_path = os.path.join(label_folder, new_image_name)

        # Rename the image
        os.rename(image_path, new_image_path)

        # Print the renaming information
        print(f"Renamed: {image_path} to {new_image_path}")
```
